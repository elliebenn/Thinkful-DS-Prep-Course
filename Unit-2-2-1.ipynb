## Basic Plot and Scatter

Matplotlic is the fundametnal plotting and visualization library for Python.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotliv inline
#With Jupyter notebook, this statement makes sure your images will always generate in the notebook 
#instead of possibly as popups, preventing a slow matplotlib compilation from giving you rendering error

plt.plot([0,1,2,3])

# You won't always need plt.show() at the end of a cell in a notebook to show the plot, 
#  but it's good practice to keep it there. 
# It will be useful to ensure that all the options and specifications you add to your plot are properly rendered. 
plt.show()
```

```python
# Make the random function consistent and replicable.
# The "seed" is a starting point for the sequence and the guarantee is that if 
#     you start from the same seed you will get the same sequence of numbers.
np.random.seed(1221)

# Make a blank data frame.
df = pd.DataFrame()

# Add a column of random numbers between 0 and 1.
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2

# When creating a data frame an index column of counts is created, counting from 0.
# Here we do a few transforms on that index to create some extra columns.
df['counts_sq'] = df.index ** 2 
df['counts_sqrt'] = np.sqrt(df.index)

df.head(10)
```

```python
plt.plot(df['rand'])
plt.show()

#By default, plot creates a line graph with the index as the x-axis. 
#   Recall that by default the index is a count starting from 0.
```

```python
#  Let's say I want to change the color to orange, expand the y-axis, 
#      label the y-axis, and give the whole thing a title.

plt.plot(df['rand'], color = 'orange')
plt.ylim([-0.5, 1.5])
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

```python
# Let's add a second series.

plt.plot(df['rand'],color = 'orange')
plt.plot(df['rand_shift'], color = 'purple')
plt.ylim([-0.1, 3.1])
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

### Scatter

```python
plt.scatter(x = df['rand'], y = df['rand_sq'])
plt.show()
```

```python
plt.scatter(
    x = df['rand'],
    y = df['rand_sq'],
    color = 'aquamarine',
    marker = 'o', s = 10
)

# s = size

plt.scatter(
    x = df['rand'],
    y = df['rand_shift'],
    color = 'goldenrod',
    marker = 'x', s = 10
)
plt.show()
```

Always feel free to look at the documentation, particularly the matplotlib.pyplot.plot() [documentation](http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot) (see the section on keyword arguments) and other methods like .ylim(), .ylabel(), and title() for more ideas. When in doubt, there's probably a way to do it.

### Pandas Plot

```python
df.plot(kind = 'scatter', x = 'counts_sq',y = 'counts_sqrt')
df.plot(kind = 'line')
plt.show()
```
