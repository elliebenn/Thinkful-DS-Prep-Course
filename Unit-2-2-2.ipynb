## Subplots

Subplots are matplotlib's way of generating multiple plots in a single figure. You break up your figure into multiple distinct areas and generate different plots in each area. This can be extremely useful for generating a series of plots and presenting them in a clearly associated way. It also can make it easy to transfer multiple visuals as a single unit. 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Make the random function consistent and replicable
np.random.seed(12210)

#Make the dataframe
df = pd.DataFrame()
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2

#Here we do a couple transforms on the index (which starts at 0)
df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)

df.head(10)
```

```python
# To put two plots together in one object as two sub plots, you need to first define the subplot using plt.subplot().
#   plt.subplot takes three arguments: two dimensions of the plot (generates subplot grid) 
#   and the identity of the subplot you're creating.

plt.figure(figsize=(10, 5))
#The plt.figure() function sets the size of the figure we're about to create, 
# taking a tuple (immutable list) with the x and y dimensions we want.

plt.subplot(1, 2, 1) # 1 high (row), 2 wide subplot (column), with Random Series in first space.
plt.plot(df['rand'], color='purple')
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(1, 2, 2)
plt.scatter(x = df['rand_sq'], y = df['rand'], color='green')
plt.ylabel('Shifted Values')
plt.title('Shifted Series')

plt.tight_layout() #Creates additional spacing around the subplots
plt.show()
```
